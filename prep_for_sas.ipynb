# data wrangling for SAS anlysis

```python
import pandas as pd
import numpy as np
```

```python
def create_incidence_columns(df):
    df['Mi_i'] = np.where((df.Incognita > 0)&(df.Arenaria == 0)&(df.Javanica == 0), 1, 0)
    df['Ma_i'] = np.where((df.Incognita == 0)&(df.Arenaria > 0)&(df.Javanica == 0), 1, 0)
    df['Mj_i'] = np.where((df.Incognita == 0)&(df.Arenaria == 0)&(df.Javanica > 0), 1, 0)

    df['MiMa_i'] = np.where((df.Incognita > 0)&(df.Arenaria > 0)&(df.Javanica == 0), 1, 0)
    df['MiMj_i'] = np.where((df.Incognita > 0)&(df.Arenaria == 0)&(df.Javanica > 0), 1, 0)
    df['MaMj_i'] = np.where((df.Incognita == 0)&(df.Arenaria > 0)&(df.Javanica > 0), 1, 0)

    df['MiMaMj_i'] = np.where((df.Incognita > 0)&(df.Arenaria > 0)&(df.Javanica > 0), 1, 0)
    return df
```

```python
def create_abundance_columns(df):
    df['Mi_a'] = np.where((df.Incognita > 0)&(df.Arenaria == 0)&(df.Javanica == 0), df.RKN, 0)
    df['Ma_a'] = np.where((df.Incognita == 0)&(df.Arenaria > 0)&(df.Javanica == 0), df.RKN, 0)
    df['Mj_a'] = np.where((df.Incognita == 0)&(df.Arenaria == 0)&(df.Javanica > 0), df.RKN, 0)

    df['MiMa_a'] = np.where((df.Incognita > 0)&(df.Arenaria > 0)&(df.Javanica == 0), df.RKN, 0)
    df['MiMj_a'] = np.where((df.Incognita > 0)&(df.Arenaria == 0)&(df.Javanica > 0), df.RKN, 0)
    df['MaMj_a'] = np.where((df.Incognita == 0)&(df.Arenaria > 0)&(df.Javanica > 0), df.RKN, 0)

    df['MiMaMj_a'] = np.where((df.Incognita > 0)&(df.Arenaria > 0)&(df.Javanica > 0), df.RKN, 0)
    return df
```

```python
df = pd.read_excel('FOR_CHART_updated_latest.xlsx')
df = df.fillna(0)
df = create_incidence_columns(df)
df = create_abundance_columns(df)
df.to_excel('survey_data_for_sas.xlsx')
df
```
